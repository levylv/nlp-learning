```python
 # pylint: disable=redefined-outer-name
  def maybe_download(filename, expected_bytes, sha256=None):
    """Download a file if not present, and make sure it's the right size."""
    local_filename = os.path.join(gettempdir(), filename)
    if not os.path.exists(local_filename):
      local_filename, _ = urllib.request.urlretrieve(url + filename,
                                                     local_filename)
    statinfo = os.stat(local_filename)

    if sha256 and _hash_file(local_filename) != sha256:
      raise Exception('Failed to verify ' + local_filename + ' due to hash '
                      'mismatch. Can you get to it with a browser?')

    if statinfo.st_size == expected_bytes:
      print('Found and verified', filename)
    else:
      print(statinfo.st_size)
      raise Exception('Failed to verify ' + local_filename +
                      '. Can you get to it with a browser?')
    return local_filename
```

```python
def _hash_file(fpath):
  hasher = hashlib.sha256()
  with open(fpath, 'rb') as fpath_file:
    for chunk in iter(lambda: fpath_file.read(65535), b''):
      hasher.update(chunk)
  return hasher.hexdigest()

  """Example of building, training and visualizing a word2vec model."""
  # Create the directory for TensorBoard variables if there is not.
  if not os.path.exists(log_dir):
    os.makedirs(log_dir)

  # Step 1: Download the data.
  # Note: Source website does not support HTTPS right now.
  url = 'http://mattmahoney.net/dc/'

filename = maybe_download(
  'text8.zip',
  31344016,
  sha256='a6640522afe85d1963ad56c05b0ede0a0c000dddc9671758a6cc09b7a38e5232')
```

```python
"""Extract the first file enclosed in a zip file as a list of words."""
with zipfile.ZipFile(filename) as f:
  data = tf.compat.as_str(f.read(f.namelist()[0])).split()
vocabulary = data
```

```python
vocabulary_size = 50000
words = vocabulary
n_words = vocabulary_size

"""Process raw inputs into a dataset."""
count = [['UNK', -1]]
count.extend(collections.Counter(words).most_common(n_words - 1))
dictionary = {word: index for index, (word, _) in enumerate(count)}
data = []
unk_count = 0
for word in words:
  index = dictionary.get(word, 0)
  if index == 0:  # dictionary['UNK']
    unk_count += 1
  data.append(index)
count[0][1] = unk_count
reversed_dictionary = dict(zip(dictionary.values(), dictionary.keys()))
data = data 
count = count
unused_dictionary = dictionary
reverse_dictionary = reversed_dictionary
```

```python
data[:7]
```

```python
vocab_size = vocabulary_size
window_size = 3
vector_dim = 300
sampling_table = sequence.make_sampling_table(vocab_size)
couples, labels = skipgrams(data, vocab_size, window_size=window_size, sampling_table = sampling_table) # 负采样生成正负样本
```

```python
couples
```

```python
word_target, word_context = zip(*couples)
word_target = np.array(word_target, dtype="int32")
word_context = np.array(word_context, dtype="int32")
labels = np.array(labels, dtype="int32")

print(couples[:10], labels[:10])
```

```python
# embedding不共享
import tensorflow.keras.backend as K
from tensorflow.keras.layers import Lambda

# build model
input_target = Input((1,)) # 1维向量
input_context = Input((1,)) # 1维向量   虽然是skipgram，但context其实还是分到不同batch了，一个batch只看一个context词

target = Embedding(vocab_size, vector_dim, name ='embedding')(input_target)
target = Reshape((vector_dim, 1))(target)
context =  Embedding(vocab_size, vector_dim, name ='W')(input_context)
bias = Embedding(vocab_size, 1, name='bias')(input_context)

output = Lambda(lambda x : K.sigmoid(K.batch_dot(x[0], x[1]) + x[2]))([context, target, bias])
model = Model(inputs=[input_target, input_context], outputs=output)
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# embedding共享
input_target = Input((1,))
input_context = Input((1,))

embedding = Embedding(vocab_size, vector_dim, input_length=1, name='embedding')
target = embedding(input_target)
target = Reshape((vector_dim, 1))(target)
context = embedding(input_context)
context = Reshape((vector_dim, 1))(context)

# # setup a cosine similarity operation which will be output in a secondary model
# similarity = merge([target, context], mode='cos', dot_axes=0)

# now perform the dot product operation to get a similarity measure
dot_product = Dot(axes=1, normalize=False)([target, context])
dot_product = Reshape((1,))(dot_product)
# add the sigmoid output layer
output = Dense(1, activation='sigmoid')(dot_product)
# create the primary training model
model = Model(inputs=[input_target, input_context], outputs=output)
model.compile(loss='binary_crossentropy', optimizer='rmsprop')
```

```python
# 引入Tensorboard
from tensorflow.keras.callbacks import TensorBoard
tbCallBack = TensorBoard(log_dir=log_dir,  # log 目录
                 histogram_freq=0,  # 按照何等频率（epoch）来计算直方图，0为不计算
#                  batch_size=32,     # 用多大量的数据计算直方图
                 write_graph=True,  # 是否存储网络结构图
                 write_grads=True, # 是否可视化梯度直方图
                 write_images=True,# 是否可视化参数
                 embeddings_freq=0, 
                 embeddings_layer_names=None, 
                 embeddings_metadata=None)
```

```python
labels = np.array(labels, dtype="int32")
def data_generator():
    while True:
        batch_size = 32
        indexs = np.random.choice(len(labels), batch_size)
        yield ([word_target[indexs], word_context[indexs]], labels[indexs])
```

```python
model.fit_generator(data_generator(), epochs=100, verbose=1, callbacks=[tbCallBack], steps_per_epoch = 1000)
```

```python
embeddings = model.get_weights()[0]
normalized_embeddings = embeddings / (embeddings**2).sum(axis=1).reshape((-1,1))**0.5

final_embeddings = normalized_embeddings
```

```python
# Step 6: Visualize the embeddings.
  # pylint: disable=missing-docstring
  # Function to draw visualization of distance between embeddings.
  def plot_with_labels(low_dim_embs, labels):
    assert low_dim_embs.shape[0] >= len(labels), 'More labels than embeddings'
    plt.figure(figsize=(18, 18))  # in inches
    for i, label in enumerate(labels):
      x, y = low_dim_embs[i, :]
      plt.scatter(x, y)
      plt.annotate(
          label,
          xy=(x, y),
          xytext=(5, 2),
          textcoords='offset points',
          ha='right',
          va='bottom')

#     plt.savefig(filename)
    plt.show()

  try:
    # pylint: disable=g-import-not-at-top
    from sklearn.manifold import TSNE
    import matplotlib.pyplot as plt

    tsne = TSNE(
        perplexity=30, n_components=2, init='pca', n_iter=5000, method='exact')
    plot_only = 500
    low_dim_embs = tsne.fit_transform(final_embeddings[:plot_only, :])
    true_labels = [reverse_dictionary[i] for i in range(plot_only)]
    plot_with_labels(low_dim_embs, true_labels)

  except ImportError as ex:
    print('Please install sklearn, matplotlib, and scipy to show embeddings.')
    print(ex)
```

```python
def most_similar(w):
    v = normalized_embeddings[unused_dictionary[w]]
    sims = np.dot(normalized_embeddings, v)
    sort = sims.argsort()[::-1]
    sort = sort[sort > 0]
    return [(reverse_dictionary[i],sims[i]) for i in sort[:10]]
```
